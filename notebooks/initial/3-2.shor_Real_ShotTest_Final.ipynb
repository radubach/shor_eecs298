```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from fractions import Fraction
from math import gcd, log2
from numpy.random import choice
from qiskit import QuantumCircuit, transpile
from qiskit_ibm_runtime import QiskitRuntimeService, SamplerV2 as Sampler
from qiskit.visualization import plot_histogram, circuit_drawer
from IPython.display import display
import time
import os

# Create directory to store result files
os.makedirs("shorReal", exist_ok=True)

# Save an IBM Cloud account, if you have IBM Cloud.
#QiskitRuntimeService.save_account(channel="ibm_cloud", token="MY_IBM_CLOUD_API_KEY", instance="MY_IBM_CLOUD_CRN")

# Please write you API token in MY_IBM_QUANTUM_TOKEN. 
# Save an IBM Quantum account.
QiskitRuntimeService.save_account(
    channel="ibm_quantum", 
    token="token key",
    overwrite=True
)
# if you don't want to save:
#service = QiskitRuntimeService(channel="ibm_quantum", token="MY_IBM_QUANTUM_TOKEN")
#service = QiskitRuntimeService()
```

```python
# Load IBM Quantum service and get the least busy backend
service = QiskitRuntimeService()
backend = service.least_busy(simulator=False, operational=True)
print(f"Using backend: {backend.name}")

# Function to construct the inverse Quantum Fourier Transform circuit
def qft_dagger(n):
    qc = QuantumCircuit(n)
    qc.name = "QFT"
    for qubit in range(n // 2):
        qc.swap(qubit, n - qubit - 1)
    for j in range(n):
        for m in range(j):
            qc.cp(-np.pi / float(2 ** (j - m)), m, j)
        qc.h(j)
    return qc

# Function to construct the controlled-U^a mod 15 circuit for given base a
def c_amod15(a, power):
    if a not in [2, 7, 8, 11, 13]:
        raise ValueError("Only a in [2, 7, 8, 11, 13] is supported for N=15.")
    U = QuantumCircuit(4)
    U.name = f"{a}^{power} mod 15"
    for _ in range(power):
        if a in [2, 13]:
            U.swap(0, 1); U.swap(1, 2); U.swap(2, 3)
        elif a in [7, 8]:
            U.swap(2, 3); U.swap(1, 2); U.swap(0, 1)
        elif a == 11:
            U.swap(2, 3); U.swap(0, 2)
        if a in [7, 11, 13]:
            for q in range(4):
                U.x(q)
    return U.to_gate().control()

# Compute Shannon entropy of measurement outcomes
def compute_entropy(counts, shots):
    probs = [v / shots for v in counts.values()]
    return -sum(p * log2(p) for p in probs if p > 0)

# Execute the QPE trial on IBM Quantum hardware
def run_qpe_trial_ibmq(a, n_count=8, shots=1024):
    qc = QuantumCircuit(n_count + 4, n_count, name=f"Shor_N15_a{a}")
    for q in range(n_count):
        qc.h(q)
    qc.x(n_count + 3)
    for q in range(n_count):
        qc.append(c_amod15(a, 2 ** q), [q] + [i + n_count for i in range(4)])
    qc.append(qft_dagger(n_count), range(n_count))
    qc.measure(range(n_count), range(n_count))

    # Draw and save the circuit
    print("Quantum Circuit:")
    fig = circuit_drawer(qc, output='mpl', fold=-1)
    fig.savefig(f"shorReal/circuit_a{a}_shots{shots}.png")
    display(fig)

    # Transpile and run on backend
    isa_circuit = transpile(qc, backend=backend)
    sampler = Sampler(mode=backend)
    job = sampler.run([isa_circuit], shots=shots)
    result = job.result()
    raw_counts = result[0].data.c.get_counts()

    # Post-process result into DataFrame
    rows = []
    for bitstring, count in raw_counts.items():
        decimal = int(bitstring, 2)
        phase = decimal / (2 ** n_count)
        frac = Fraction(phase).limit_denominator(15)
        r = int(frac.denominator)
        rows.append([bitstring.zfill(n_count), count, phase, f"{frac.numerator}/{frac.denominator}", r])

    df = pd.DataFrame(rows, columns=["Bitstring", "Counts", "Phase", "Fraction", "Candidate r"])
    return df, raw_counts

# Try to compute nontrivial factor using given candidate r
def try_factorization(a, N, r):
    if r % 2 != 0:
        print(f"  Rejected: r = {r} is not even.")
        return None
    if pow(a, r, N) != 1:
        print(f"  Rejected: a^{r} mod N  1")
        return None
    x1 = pow(a, r // 2) - 1
    x2 = pow(a, r // 2) + 1
    guess1 = gcd(x1, N)
    guess2 = gcd(x2, N)
    print(f"  Trying gcd(a^{r//2} - 1, N) = gcd({x1}, {N}) = {guess1}")
    print(f"  Trying gcd(a^{r//2} + 1, N) = gcd({x2}, {N}) = {guess2}")
    if 1 < guess1 < N:
        print(f"  Success: Nontrivial factor found: {guess1}")
        return guess1
    if 1 < guess2 < N:
        print(f"  Success: Nontrivial factor found: {guess2}")
        return guess2
    print("  Failed: No nontrivial factor found with this r.")
    return None

# Experiment configuration
N = 15
n_count = 8
valid_a_values = [2, 7, 8, 11, 13]
shots_list = [4, 8, 16, 32, 64, 128, 256, 512, 1024, 2048]
summary_rows = []

# Main experimental loop
for shots in shots_list:
    print(f"\n=== Running trials with shots = {shots} ===")
    total_time = 0
    total_entropy = 0
    total_peak_conf = 0
    total_phase_std = 0
    success_trials = 0
    total_attempts = []

    for trial in range(3):
        a = int(choice(valid_a_values))
        print(f"\nTrial {trial+1}: Trying a = {a}")
        start = time.time()
        df, counts = run_qpe_trial_ibmq(a, n_count, shots)
        elapsed = time.time() - start
        total_time += elapsed

        df.to_csv(f"shorReal/phase_table_shots{shots}_trial{trial+1}.csv", index=False)

        # Show result table and histogram
        print("\nPhase Estimation Result Table:")
        display(df)

        print("\nMeasurement Histogram:")
        fig = plot_histogram(counts, title=f"Histogram (a={a}, shots={shots})")
        fig.savefig(f"shorReal/histogram_a{a}_shots{shots}_trial{trial+1}.png")
        plt.show()

        # Try top 2 most frequent r candidates
        df_sorted = df.sort_values(by="Counts", ascending=False).head(2)
        tried_r = set()
        for _, row in df_sorted.iterrows():
            r = int(row["Candidate r"])
            if r in tried_r:
                continue
            tried_r.add(r)
            print(f"Trying candidate r = {r} from bitstring {row['Bitstring']} (phase  {row['Phase']:.4f})")
            if try_factorization(a, N, r):
                print(f"  Trial {trial+1}: Success with a = {a}, r = {r}")
                total_attempts.append(1)
                success_trials += 1
                break
        else:
            print(f"  Trial {trial+1}: Failed to factor using top 2 candidates.")
            total_attempts.append(2)

        # Record metrics for this trial
        total_entropy += compute_entropy(counts, shots)
        total_peak_conf += max(counts.values()) / shots
        phases = [float(p) for p in df["Phase"] for _ in range(int(df.loc[df["Phase"] == p, "Counts"].values[0]))]
        total_phase_std += np.std(phases) if phases else 0

    # Aggregate trial results
    avg_attempts = np.mean(total_attempts)
    avg_time = total_time / 3
    success_rate = success_trials / 3
    failure_rate = 1 - success_rate
    avg_entropy = total_entropy / 3
    avg_conf = total_peak_conf / 3
    avg_std = total_phase_std / 3

    summary_rows.append((shots, avg_attempts, success_rate, failure_rate, avg_std, avg_conf, avg_entropy, avg_time))

# Final result table and CSV export
summary_df = pd.DataFrame(summary_rows, columns=[
    "Shots", "Average Attempts", "Success Rate", "Failure Rate",
    "Avg Phase Std", "Avg Peak Confidence", "Avg Entropy", "Avg Time (s)"
])
summary_df.to_csv("shorReal/shor_ibmq_detailed_metrics.csv", index=False)

# Plot and save graphs for each metric
metrics = [
    ("Average Attempts", "o"),
    ("Success Rate", "s"),
    ("Failure Rate", "x"),
    ("Avg Phase Std", "D"),
    ("Avg Peak Confidence", "^"),
    ("Avg Entropy", "v"),
    ("Avg Time (s)", "h")
]

for metric, marker in metrics:
    plt.figure(figsize=(8, 5))
    plt.plot(summary_df["Shots"], summary_df[metric], marker=marker)
    plt.title(f"{metric} vs Shots (IBM Q Hardware)")
    plt.xlabel("Shots")
    plt.ylabel(metric)
    plt.grid(True)
    plt.savefig(f"shorReal/{metric.replace(' ', '_')}_vs_Shots.png")
    plt.show()
```
